# Data Visualization

# 1) Number of Terrorist Attacks each Year

```python
year=data['iyear'].unique()
years_count= data['iyear'].value_counts(dropna = False).sort_index()
plt.figure(figsize=(15,10)) 
sns.barplot(x=year, y=years_count, palette='rocket') 
plt.xticks(rotation=50)
plt.xlabel("Attacking Year", fontsize=20) 
plt.ylabel("Number of Attacks Each Year",fontsize=20) 
plt.title("Attacks In Years", fontsize=22)
plt.show()
```

# 2) Top 10 terrorist group attack

```python
data['gname'].value_counts().to_frame().drop("Unknown").head(10).plot(kind="bar",color="yellow", figsize=(20,10)) 
plt.title("Top 10 terrorist group attack", fontsize=25)
plt.xlabel("terrorist group name", fontsize=20) 
plt.ylabel("Attack count", fontsize=20)
plt.show()
```

# 3) Number Of Terrorist Activities Each Year

```python
plt.subplots(figsize=(15, 6))
sns.countplot(x='iyear', data=data, palette='RdYlGn_r', edgecolor=sns.color_palette("YlOrBr", 10))
plt.xticks(rotation=45)
plt.title('Number Of Terrorist Activities Each Year')
plt.show()
```

# 4) Terrorist Activities by Region in each Year 

```python
pd.crosstab(data.iyear, data.region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

# 5) Top 10 most affected terriost attack regions

```python
plt.figure(figsize=(25, 12))
sns.barplot(x=data["region"].value_counts()[:10].index, y=data["region"].value_counts()[:10].values, edgecolor='k')
plt.xlabel('Region', fontweight='bold', fontsize=18)
plt.ylabel('Counts', fontweight='bold', fontsize=18)
plt.show()
```

# 6) Top 10 most affected terriost attack regions

# 7) Top 10 most affected terriost attack States

```python
plt.figure(figsize=(25, 12))
sns.barplot(x=data["provstate"].value_counts()[:10].index, y=data["provstate"].value_counts()[:10].values, edgecolor='k')
plt.xlabel('States', fontweight='bold', fontsize=18)
plt.ylabel('Counts', fontweight='bold', fontsize=18)
plt.title('Top 10 most affected terriost attack States',fontweight='bold',fontsize=25)
plt.show()
```

# 8) Top 10 most attacked cities

```python
plt.figure(figsize=(25,12))
sns.barplot(x=data["city"].value_counts()[:10].index,y=data["city"].value_counts()[:10].values,edgecolor='k')
plt.xlabel('Cities',fontweight='bold',fontsize=18)
plt.ylabel('Count',fontweight='bold',fontsize=18)
plt.title('Top 10 most affected terriost attack in Cities',fontweight='bold',fontsize=25)
plt.xticks(rotation=60)
plt.show()
```

# 9) Types of weapon used for terrorist activities

```python
plt.figure(figsize=(25,12))
sns.countplot(x=data["weaptype1_txt"],order=data["weaptype1_txt"].value_counts().index,edgecolor='k')
plt.xlabel('Weapon Type',fontweight='bold',fontsize=18)
plt.ylabel('Number Weapons',fontweight='bold',fontsize=18)
plt.title('Type of Weapons',fontweight='bold',fontsize=25)
```
